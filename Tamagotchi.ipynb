```python
class Tamagotchi:
    def __init__(self, nombre, color, salud=100, felicidad=100, energia=100):
        self.nombre = nombre
        self.color = color
        self.salud = salud
        self.felicidad = felicidad
        self.energia = energia

    def jugar(self):
        self.felicidad += 10
        self.salud -= 5
        print(f"{self.nombre} jugó")

    def comer(self):
        self.felicidad += 5
        self.salud += 10
        print(f"{self.nombre} comió")

    def curar(self):
        self.salud += 20
        self.felicidad -= 5
        print(f"{self.nombre} fue curado")

    def mostrar_estado(self):
        print(f"{self.nombre} → Salud: {self.salud}, Felicidad: {self.felicidad}, Energía: {self.energia}")
```

```python
class Persona:
    def __init__(self, nombre, apellido, tamagotchi):
        self.nombre = nombre
        self.apellido = apellido
        self.tamagotchi = tamagotchi

    def jugar_con_tamagotchi(self):
        print(f"{self.nombre} está jugando con {self.tamagotchi.nombre}")
        self.tamagotchi.jugar()

    def darle_comida(self):
        print(f"{self.nombre} le da comida a {self.tamagotchi.nombre}")
        self.tamagotchi.comer()

    def curarlo(self):
        print(f"{self.nombre} cura a {self.tamagotchi.nombre}")
        self.tamagotchi.curar()
```

```python
# Crear Tamagotchi
tama1 = Tamagotchi("Pikachu", "Amarillo")

# Crear Persona
persona1 = Persona("Wil", "Al", tama1)

# Interacciones
persona1.darle_comida()
persona1.jugar_con_tamagotchi()
persona1.curarlo()

# Ver estado final
tama1.mostrar_estado()
```
