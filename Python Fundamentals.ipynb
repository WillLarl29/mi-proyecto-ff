# Python Fundamentals / Ejercicios Listas, Tuplas y diccionarios (Core)

### 1. Carga de Datos:

```python
ventas = [
    {"fecha": "2024-01-01", "producto": "Laptop", "cantidad": 2, "precio": 1500.0},
    {"fecha": "2024-01-01", "producto": "Mouse", "cantidad": 5, "precio": 20.0},
    {"fecha": "2024-01-02", "producto": "Laptop", "cantidad": 1, "precio": 1500.0},
    {"fecha": "2024-01-02", "producto": "Teclado", "cantidad": 3, "precio": 45.0},
    {"fecha": "2024-01-03", "producto": "Mouse", "cantidad": 2, "precio": 20.0}
]
```

### 2. Cálculo de Ingresos Totales

```python
ingresos_totales = 0

for venta in ventas:
    ingresos_totales += venta["cantidad"] * venta["precio"]

print("Ingresos totales:", ingresos_totales)
```

### 3. Análisis del Producto Más Vendido:

```python
ventas_por_producto = {}

for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]
    
    if producto not in ventas_por_producto:
        ventas_por_producto[producto] = 0
    
    ventas_por_producto[producto] += cantidad

# producto más vendido
producto_mas_vendido = max(ventas_por_producto, key=ventas_por_producto.get)

print("Ventas por producto:", ventas_por_producto)
print("Producto más vendido:", producto_mas_vendido)
```

### 4. Promedio de Precio por Producto

```python
precios_por_producto = {}

for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]
    precio = venta["precio"]
    
    if producto not in precios_por_producto:
        precios_por_producto[producto] = [0, 0]  # [suma_precios, cantidad_total]
    
    precios_por_producto[producto][0] += precio * cantidad
    precios_por_producto[producto][1] += cantidad

# calcular promedios
for producto, valores in precios_por_producto.items():
    suma_precios, cantidad_total = valores
    promedio = suma_precios / cantidad_total
    print(f"{producto} → Precio promedio: {promedio:.2f}")
```

### 5. Ventas por Día:

```python
ingresos_por_dia = {}

for venta in ventas:
    fecha = venta["fecha"]
    ingreso = venta["cantidad"] * venta["precio"]
    
    if fecha not in ingresos_por_dia:
        ingresos_por_dia[fecha] = 0
    
    ingresos_por_dia[fecha] += ingreso

print("Ingresos por día:", ingresos_por_dia)
```

### 6. Representación de Datos:

```python
resumen_ventas = {}

for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]
    ingreso = cantidad * venta["precio"]
    
    if producto not in resumen_ventas:
        resumen_ventas[producto] = {
            "cantidad_total": 0,
            "ingresos_totales": 0,
            "precio_acumulado": 0
        }
    
    resumen_ventas[producto]["cantidad_total"] += cantidad
    resumen_ventas[producto]["ingresos_totales"] += ingreso
    resumen_ventas[producto]["precio_acumulado"] += ingreso

# calcular precio promedio
for producto, datos in resumen_ventas.items():
    datos["precio_promedio"] = datos["precio_acumulado"] / datos["cantidad_total"]
    del datos["precio_acumulado"]  # opcional

print("Representación de Datos::")
for producto, datos in resumen_ventas.items():
    print(producto, "→", datos)
```
